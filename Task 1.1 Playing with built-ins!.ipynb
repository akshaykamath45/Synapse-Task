```python
modern_family = ['CLaiRe_DunPhY', 'PHiL_dUnpHY', 'GLoRiA_PriTCheTt', 'CaMErOn_TuCKEr','StELLa']
indices=[]
characters=[]
y= list(enumerate(modern_family))

for i in range(len(modern_family)):
  indices.append(y[i][0])
  characters.append(y[i][1])

print("Characters are: ",characters)
print("Indices are: ",indices)
```

```python
for i in range(len(characters)):
   characters[i] = characters[i].lower()
print("Lower Characters are: ",characters)

for i in range(len(characters)):
   characters[i] = characters[i].replace("_","-")
print("Replaced Characters are: ",characters)
```

```python
temp = list(map(lambda x: len(x), characters))
temp
```

```python
new =list(zip(temp,indices))
print(new)
```
